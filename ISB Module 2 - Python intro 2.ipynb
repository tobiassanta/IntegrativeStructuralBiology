# Slicing

```python
s       = "hello"
l       = [1, 2, 3, 4, 5]

print(s[1:4])
print(s[1:5:3])
print(s[4:])
print(s[1:])
print(s[:])

print(l[1:4])
print(l[1:5:3])
print(l[4:])
print(l[1:])
print(l[:])
```

# Exercise 1

```python
quote = '''Horse sense is the thing a horse has which keeps it
        from betting on people. W.C. Fields'''

# List of words
quote_list = quote.split()
print(quote_list)

# Index of "Horse"
quote_list.index('Horse')

# Sorting the list
quote_list.sort()

# Printing the first three elements of the sorted list
print(quote_list[:3])

# Reversing the order of the list
quote_list_reversed = quote_list[::-1]
print(quote_list_reversed)
```
