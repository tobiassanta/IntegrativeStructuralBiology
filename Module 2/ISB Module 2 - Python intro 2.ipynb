# Slicing

```python
s       = "hello"
l       = [1, 2, 3, 4, 5]

print(s[1:4])
print(s[1:5:3])
print(s[4:])
print(s[1:])
print(s[:])

print(l[1:4])
print(l[1:5:3])
print(l[4:])
print(l[1:])
print(l[:])
```

# Exercise 1

```python
quote = '''Horse sense is the thing a horse has which keeps it
        from betting on people. W.C. Fields'''

# List of words
quote_list = quote.split()
print(quote_list)

# Index of "Horse"
quote_list.index('Horse')

# Sorting the list
quote_list.sort()

# Printing the first three elements of the sorted list
print(quote_list[:3])

# Reversing the order of the list
quote_list_reversed = quote_list[::-1]
print(quote_list_reversed)
```

# Exercise 4

```python
num_list = [6, 9, 4, 8, 7, 1, 2]

# While loop that prints each element
i = 0
while i < len(num_list):
    print(num_list[i])
    i += 1
```

```python
# For loop that prints each element
for num in num_list:
    print(num)
```

```python
# Print all elements that are larger than their left neighbor
for i in range(len(num_list)):
    if num_list[i-1] < num_list[i]:
        print(num_list[i])

# or one could do

for i, num in enumerate(num_list):
    if i > 0 and num > num_list[i-1]:
        print(num)
```
