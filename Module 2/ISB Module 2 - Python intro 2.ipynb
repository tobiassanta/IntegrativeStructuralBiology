# Exercise 1

```python
# 1) Defining quote
quote = '''Horse sense is the thing a horse has which keeps 
           it from betting on people. W.C. Fields'''
```

```python
# 2) List of words
quote_list = quote.split()
print(quote_list)
```

```python
# 3) Index of "Horse"
print(quote_list.index('Horse'))
```

```python
# 4) Sorting the list
quote_list.sort()
print(quote_list)
```

```python
# 5) Printing the first three elements of the sorted list
print(quote_list[:3])
```

```python
# 6) Reversing the order of the list
quote_list_reversed = quote_list[::-1]
print(quote_list_reversed)
```

```python
# 4) While loop that prints each element in the list
i = 0
while i < len(word_list):
    print(word_list[i])
    i += 1
```

# Exercise 4

```python
# 1) Make list
num_list = [6, 9, 4, 8, 7, 1, 2]
```

```python
# 2) While loop that prints each element
i = 0
while i < len(num_list):
    print(num_list[i])
    i += 1
```

```python
# 3) For loop --||--
for num in num_list:
    print(num)
```

```python
# 4) Print every element that is larger than its left neighbor
for i, num in enumerate(num_list):
    if num > num_list[i-1]:
        print(num)
```
