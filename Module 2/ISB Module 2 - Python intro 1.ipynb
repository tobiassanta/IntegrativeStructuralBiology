# Recap on datatypes

1) Types of numbers: Integer (42), float (42.0), complex (42.0 + 3.0j)

2) What are booleans used for?: They are used in conditional statements (IF, ELSE), loops (WHILE etc.), logical operations (AND, OR, NOT) to determine which code blocks execute.

3) How do you create an empty string?: Using empty doublequotes (" "), empty single quotes (' '), empty triple quotes (""" """ or ''' '''), or empty str()

4) What does "\n" mean?: Newline

5) What does * do for strings?: Repetion of the given string

6) Can you add a string to a number?: No, adding numbers to strings gives an error, however, you can concatenate two strings using addition of strings.

# Exercise 2

```python
import random

x = random.random() * 6
if x < 1:
    print('Dice Roll = 1')
elif x < 2:
    print('Dice Roll = 2')
elif x < 3:
    print('Dice Roll = 3')
elif x < 4:
    print('Dice Roll = 4')
elif x < 5:
    print('Dice Roll = 5')
else:
    print('Dice Roll = 6')
```
